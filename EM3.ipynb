## Clean Month


```python
recweek.birthmonth.unique()
```

```python
#write code here
month_list=["","January","February","March","April","May","June","July","August","September","October","November","December"]
def clean_month(value):
    # There is no 0th month
    try:
        month_number =int(value)
        return month_list[month_number]
    except ValueError:
        months=month_list[1:]
        if value in month_list:
            return value
        month_series = pd.Series(month_list)
        return month_series[month_series.str.startswith(value)].values[0]
    
    
```

```python
recweek["birthmonth"]=recweek.birthmonth.apply(clean_month)
```

```python
# Clean Year



# dump recweek data
recweek
```

```python
recweek.birthyear.unique()
```

```python
#write code here
def clean_year(value):
    # if value is not 4 digits
    if value/1000 <1:
        return 1900+ value
    return value
```

```python
recweek["birthyear"]= recweek.birthyear.apply(clean_year)
recweek
```

## Clean Courses

```python
# Clean Course




# dump recweek data
recweek
```

```python
recweek.course.unique()
```

```python
course_map={
    "LM":"Legal Management",
    "CTM": "Communications Technology",
    "ITE": "IT Entrepreneurship",
    "MGT": "Management",
    "ME": "Management Engineering",
}

#write code here
def clean_courses(value):
    # Information Technology Entrepreneurship, shorten it.
    if value=="Information Technology Entrepreneurship":
        return "BS IT Entrepreneurship"
    if value in course_map.values():
        return "BS "+ value
    if value.startswith("BS "):
        value = value.replace("BS ","")
    return "BS "+course_map[value]
```

```python
recweek["course"]= recweek.course.apply(clean_courses)
recweek
```
