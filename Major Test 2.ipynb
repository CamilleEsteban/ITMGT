### Data Visualization

Specifications:
* Figure consists of 1 row by 2 columns
* Figure size is 12 inches x 20 inches
* Amounts must be in thousands (so you'll have to divide by 1000)

Useful links:  
https://stackoverflow.com/questions/10998621/rotate-axis-text-in-python-matplotlib
https://stackoverflow.com/questions/10998621/rotate-axis-text-in-python-matplotlib

```python
import matplotlib.pyplot as plt
```

```python
df_country_sorted[2019]=df_country_sorted[2019]/1000
df_country_sorted
```

```python
df_continent_sorted[2019]= df_continent_sorted[2019]/1000
df_continent_sorted
```

```python
df_USA_sorted= df_country[df_country.Country=="U.S.A."][months].transpose()
df_USA_sorted.rename(columns={66:"remittances"},inplace=True)
df_USA_sorted.sort_values(by="remittances",inplace=True,ascending=False)
```

```python
df_USA_sorted["remittances"]= df_USA_sorted["remittances"]/1000
```

```python
df_USA_sorted
```

```python
# Make a 1x3 Figure
%matplotlib inline 
fig, axes= plt.subplots(3,1, figsize=(12,20))
# First Graph
# Horizontal Bar Graph
# Bar color: Green ("Green")
df_country_topten=df_country_sorted.head(10).sort_values(by=2019,ascending=True)
axes[0].barh(y=df_country_topten["Country"],width=df_country_topten[2019],color="Green")
axes[0].set_xlabel("Remittances(in Thousands)")
axes[0].set_ylabel("Country")
axes[0].set_title("Top 10 Sources of Cash Remittances")
# Second Graph
# Vertical Bar
# Bar Color: Orange ("Orange")
df_continent_topten=df_continent_sorted.head(10).sort_values(by=2019,ascending=False)
axes[1].bar(height=df_continent_topten[2019],x=df_continent_topten["Country"],color="Orange")
axes[1].set_ylabel("Remittances(in Thousands)")
axes[1].set_xlabel("Country")
axes[1].set_title("Top 10 Sources of Cash Remittances")

# Third Graph
axes[2].bar(height=df_USA_sorted["remittances"],x=df_USA_sorted.index,color="r")
axes[2].set_ylabel("Remittances(in Thousands)")
axes[2].set_xlabel("Month")
axes[2].set_title("Monthly Remittances")
# show plot

plt.show()
```
